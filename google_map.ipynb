```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

# Function to extract company information
def extract_company_info(driver, link_url):
    try:
        driver.get(link_url)
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")

        company_name_element = soup.find('h1', class_='DUwDvf lfPIob')
        company_name = company_name_element.text.strip() if company_name_element else "Company name not found."

        address_element = soup.find('button', {'data-item-id': 'address'})
        address = address_element.find('div', class_='Io6YTe').text.strip() if address_element else "Address not found."

        website_element = soup.find('a', {'data-item-id': 'authority'})
        website_link = website_element['href'] if website_element else "Website link not found."
        
        # Scrape the phone number
        phone_element_1 = soup.find('button', {'jsaction': 'pane.wfvdle11;clickmod:pane.wfvdle11;focus:pane.focusTooltip;blur:pane.blurTooltip'})
        phone_number_1 = phone_element_1['aria-label'].replace('Phone: ', '').strip() if phone_element_1 else "Phone number 1 not found."

        phone_element_2 = soup.find('button', {'jsaction': 'pane.wfvdle9;clickmod:pane.wfvdle9;focus:pane.focusTooltip;blur:pane.blurTooltip'})
        phone_number_2 = phone_element_2['aria-label'].replace('Phone: ', '').strip() if phone_element_2 else "Phone number 2 not found."

        return company_name, website_link, phone_number_1, phone_number_2, address, link_url
    except Exception as e:
        # If an exception occurs, handle it here and store the error messages
        return "Error", "Error", "Error", "Error", "Error", link_url

# Set up the webdriver
#driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Base URL for the product search
#base_url = 'https://www.google.com/maps/search/software+house+lahore+pakistan/@31.4837737,74.2133143,11.75z?entry=ttu'

# Fetch the HTML content from the URL
#driver.get(base_url)
driver.implicitly_wait(20)
# Scroll down to the bottom of the page to load more content
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

html_content = driver.page_source
soup = BeautifulSoup(html_content, "html.parser")

company_divs = soup.find_all("div", class_="Nv2PK tH5CWc THOPZb")

# Lists to store company information
company_names = []
website_links = []
phone_numbers_1 = []
phone_numbers_2 = []
addresses = []
google_links = []

for company_div in company_divs:
    anchor_element = company_div.find('a', class_='hfpxzc')
    if anchor_element is not None:
        link_url = anchor_element['href']
        #print("Google Link:", link_url)

        # Extract company information using the helper function
        company_name, website_link, phone_number_1, phone_number_2, address, google_link = extract_company_info(driver, link_url)

        # Append the data to the respective lists
        company_names.append(company_name)
        website_links.append(website_link)
        phone_numbers_1.append(phone_number_1)
        phone_numbers_2.append(phone_number_2)
        addresses.append(address)
        google_links.append(link_url)
    else:
        print("Link not found for this company.")

# Create a DataFrame to store the data
data = {
    'Company Name': company_names,
    'Website Link': website_links,
    'Phone Number_1': phone_numbers_1,
    'Phone Number_2': phone_numbers_2,
    'Address': addresses,
    'Google Link': google_links
}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)

# Close the webdriver
driver.quit()
```
